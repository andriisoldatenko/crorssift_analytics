```python
BASE_URL = "https://www.crossfit.com/heroes"
```

```python
resp = requests.get(BASE_URL)
resp.raise_for_status()
parsed_html = BeautifulSoup(resp.text, "html.parser")
```

```python
sections = parsed_html.body.findAll("section")
```
